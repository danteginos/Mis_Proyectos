* 2.- Crear Histogramas

```python
df["happiness_score"].max()
```

```python
# Histograma para columna happiness_score

plt.hist(df['happiness_score'], bins=10, color='cyan', edgecolor='black', alpha=0.7)
plt.title("Histograma Puntuación de Felicidad", fontsize=14, fontweight='bold')
plt.xlabel("happines_score", fontweight='bold')
plt.ylabel("Frecuencia", fontweight='bold')
plt.grid(True, linewidth=0.5)
plt.show()
```

```python
# Histograma para columna 'economy_(gdp_per_capita)'

plt.hist(df['economy_(gdp_per_capita)'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histograma Economia(PIB per Capita)", fontsize=14, fontweight='bold')
plt.xlabel("economy_(gdp_per_capita)", fontweight='bold')
plt.ylabel("Frecuencia", fontweight='bold')
plt.grid(True, linewidth=0.5)
plt.show()
```

```python
# Histograma para columna 'health_(life_expectancy)'

plt.hist(df['health_(life_expectancy)'], bins=10, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title("Histograma Salud (Esperanza de vida)", fontsize=14, fontweight='bold')
plt.xlabel('health_(life_expectancy)', fontweight='bold')
plt.ylabel("Frecuencia", fontweight='bold')
plt.grid(True, linewidth=0.5)
plt.show()
```

```python
# Histograma para columna 'freedom'

plt.hist(df['freedom'], bins=10, color='lightsalmon', edgecolor='black', alpha=0.7)
plt.title("Histograma Libertad para tomar Decisiones en la vida", fontsize=14, fontweight='bold')
plt.xlabel('freedom', fontweight='bold')
plt.ylabel("Frecuencia", fontweight='bold')
plt.grid(True, linewidth=0.5)
plt.show()
```

* 3.- Crear Boxplots:

```python
# Boxplt para la columna 'happiness_score'
prop_caja = dict(facecolor='skyblue')
plt.boxplot(df['happiness_score'], patch_artist=True, boxprops=prop_caja)
plt.title("Bloxplot Puntuación de Felicidad", fontsize=14, fontweight='bold')
plt.ylabel("Valores", fontweight='bold')
plt.show()
```

```python
# Boxplot para la columna 'economy_(gdp_per_capita)'
prop_caja = dict(facecolor='blue')
plt.boxplot(df['economy_(gdp_per_capita)'], patch_artist=True, boxprops=prop_caja)
plt.title("Bloxplot Economia(PIB per Capita)", fontsize=14, fontweight='bold')
plt.ylabel("Valores", fontweight='bold')
plt.show()
```

```python
# Boxplot para la columna ''health_(life_expectancy)''
prop_caja = dict(facecolor='green')
plt.boxplot(df['health_(life_expectancy)'], patch_artist=True, boxprops=prop_caja)
plt.title("Bloxplot Salud (Esperanza de vida)", fontsize=14, fontweight='bold')
plt.ylabel("Valores", fontweight='bold')
plt.show()
```

```python
# Boxplot para la columna 'freedom'
prop_caja=dict(facecolor='salmon')
plt.boxplot(df['freedom'], patch_artist=True, boxprops=prop_caja)
plt.title("Bloxplot Libertad para tomar Decisiones en la vida", fontsize=14, fontweight='bold')
plt.ylabel("Valores", fontweight='bold')
plt.show()
```

* 4.- Análisis Combinado:

```python
# # Crear la figura y los subplots.
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# Histograma para columna 'happiness_score'
axs[0].hist(df['happiness_score'], bins=10, color='cyan', edgecolor='black', alpha=0.7)
axs[0].set_title('Histograma Puntuación Felicidad', fontsize=14, fontweight='bold')
axs[0].set_xlabel("happines_score", fontweight='bold')
axs[0].set_ylabel("Frecuencia", fontweight='bold')
axs[0].grid(True, linewidth=0.5)


# Boxplt para la columna 'happiness_score'
prop_caja = dict(facecolor='skyblue')
axs[1].boxplot(df['happiness_score'], patch_artist=True, boxprops=prop_caja)
axs[1].set_title("Bloxplot Puntuación Felicidad", fontsize=14, fontweight='bold')
axs[1].set_xticklabels(['happines_score'], fontweight='bold')
axs[1].set_ylabel("Valores", fontweight='bold')

# Ajustar el espacio entre gráficos.
plt.tight_layout()
plt.show()
```

* En el gráfico Histograma se verifica que la mayoria de los paises tienen puntuaciones de felicidad agrupados alrededor de un valor medio entre 4 y 6.
* Hay una pequeña cantidad de paises con puntuación de felicidad baja.

* En el gráfico boxplot se verifica que no hay valores Outliers; datos mas alla de los bigotes que indicarian paises con puntuación de felicidad extremadamente alta o muy bajas.
* la mediana esta inclinada hacia abajo,indica que hay mas datos concentrados en la parte inferior.







```python
# Crear la figura y los subplots.
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# Histograma para columna 'economy_(gdp_per_capita)'.
axs[0].hist(df['economy_(gdp_per_capita)'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
axs[0].set_title('Histograma Economia', fontsize=14, fontweight='bold')
axs[0].set_xlabel("economy_(gdp_per_capita)", fontweight='bold')
axs[0].set_ylabel("Frecuencia", fontweight='bold')
axs[0].grid(True, linewidth=0.5)

# Boxplot para la columna 'economy_(gdp_per_capita)'
prop_caja = dict(facecolor='blue')
axs[1].boxplot(df['economy_(gdp_per_capita)'], patch_artist=True, boxprops=prop_caja)
axs[1].set_title("Bloxplot Economia", fontsize=14, fontweight='bold')
axs[1].set_xticklabels(['economy_(gdp_per_capita)'], fontweight='bold')
axs[1].set_ylabel("Valores", fontweight='bold')

# Ajustar el espacio entre gráficos.
plt.tight_layout()
plt.show()
```

* En el gráfico Histograma se puede ver que las mayores frecuencias estan alrededor de 1.0 y 1.25, esto sugiere que muchos paises tienen un PIB per capita en ese rango.
* En el gráfico Boxplot se puede ver que la mediana esta en el rango de 0.75 y 1.0, es decir que el 50% de paises muestra mayor desarrollo economico.
* El rango de 1.25 a 1.75 se encuentran los paises muy ricos.
* Menor a 0.25 se encuentran los paises mas pobres, con economias fragiles.
* no se ven datos outliers en gráfico boxplot.


```python
# Crear la figura y los subplots.
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# Histograma para columna 'health_(life_expectancy)'.
axs[0].hist(df['health_(life_expectancy)'], bins=10, color='lightgreen', edgecolor='black', alpha=0.7)
axs[0].set_title('Histograma Salud', fontsize=14, fontweight='bold')
axs[0].set_xlabel("health_(life_expectancy)", fontweight='bold')
axs[0].set_ylabel("Frecuencia", fontweight='bold')
axs[0].grid(True, linewidth=0.5)

# Boxplot para la columna 'health_(life_expectancy)'
prop_caja = dict(facecolor='green')
axs[1].boxplot(df['health_(life_expectancy)'], patch_artist=True, boxprops=prop_caja)
axs[1].set_title("Bloxplot Salud", fontsize=14, fontweight='bold')
axs[1].set_xticklabels(['health_(life_expectancy)'], fontweight='bold')
axs[1].set_ylabel("Valores", fontweight='bold')

# Ajustar el espacio entre gráficos.
plt.tight_layout()
plt.show()
```

* En el gráfico Histograma se verifica entre el rago de 0.6 a 1.0 estan las frecuencias mas altas, es decir distribución de paises con mayor esperanza de vida.
* Los valores de la frecuencia disminuyen hacia la izquierda con valores mas bajos de esperanza de vida.
* La mayoria de paises tienen una esperanza de vida relativamente alta

* En el gráfico boxplot, el bigote inferios es mas largo, hay una mayor variabilidad de datos en el extremo inferior
* Ausencia de outliers en los extremos
* LA mediana se encuentra en 0.6 y 0.8, la mitad de los paises tienen esperanza de vida sobre este valor

```python
# Crear la figura y los subplots.
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# Histograma para columna 'freedom'.
axs[0].hist(df['freedom'], bins=10, color='lightsalmon', edgecolor='black', alpha=0.7)
axs[0].set_title('Histograma Libertad', fontsize=14, fontweight='bold')
axs[0].set_xlabel("freedom", fontweight='bold')
axs[0].set_ylabel("Frecuencia", fontweight='bold')
axs[0].grid(True, linewidth=0.5)

# Boxplot para la columna 'freedom'
prop_caja = dict(facecolor='salmon')
axs[1].boxplot(df['freedom'], patch_artist=True, boxprops=prop_caja)
axs[1].set_title("Bloxplot Libertad", fontsize=14, fontweight='bold')
axs[1].set_xticklabels(['freedom'], fontweight='bold')
axs[1].set_ylabel("Valores", fontweight='bold')

# Ajustar el espacio entre gráficos.
plt.tight_layout()
plt.show()
```

* Se verifica en el gráfico que la frecuencia aumenta hacia la derecha con la mayoria de los paises teniendo valores altos de libertad.
* Alrededor de 0.4 se encuentra el valor de frecuencia mas alto correpondiente  a una mayor concentracion de paises.

* En el gráfico boxplot, la mediana se encuentra en el rango de 0.4 y 0.5
* Ausencias de valores outliers  
* El bigote del extremo inferior es mas largo, hay una mayor variabilidad de los datos.
