Código con Errores

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""
# Se debe importar la biblioteca StringIO
from io import StringIO

# Cargar los datos en un DataFrame de Pandas
df = pd.read_csv(StringIO(data))

# Al eliminar metodo compat se puede crear el data set, al parecer ya no es compatible con la version de phyton

```

```python
df.info(6)
```

```python
# Convertir la columna 'fecha' a formato datetime
#df['fecha'] = pd.to_datetime(df['fechas'])

# No existe columna 'fechas'
df['fecha'] = pd.to_datetime(df['fecha'])
```

```python
df.dtypes
```

```python
df.head(6)
```

```python
# Gráfico de líneas de las ventas
# Un gráfico de líneas se utiliza para mostrar datos a lo largo del tiempo o en un orden secuencial.

plt.figure(figsize=(8, 6))
plt.plot(df['fecha'], df['ventas'])
plt.title("Ventas Diarias", fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.xlabel("Fecha", fontweight='bold')
plt.ylabel("Ventas",fontweight='bold')
plt.legend()
plt.show()
```

```python
# Crear un gráfico de dispersión
plt.scatter(df['fecha'], df['ventas'])
plt.title("Ventas Diarias", fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.xlabel("Fecha", fontweight='bold')
plt.ylabel("Ventas", fontweight='bold')
plt.show()

```

```python
# Crear un histograma de las ventas
plt.hist(df['ventas'], bins=6, color='skyblue', edgecolor='black')
plt.title("Distribución de Ventas", fontsize=14, fontweight='bold')
plt.xlabel("Ventas", fontweight='bold')
plt.ylabel("Frecuencia", fontweight='bold')
plt.show()
```
