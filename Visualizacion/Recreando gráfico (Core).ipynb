Reproducir un Gráfico Completo

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Cargar el dataset
df = sns.load_dataset('tips')
```

```python
# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(12, 6))

# Se cambia el color de fondo del gráfico a color lightgray
ax.set_facecolor('lightgray')

# Gráfico de dispersión con cuadrícula, leyenda, anotaciones y flechas
# Se aplico el parametro : palette, para cambiar el color de los puntos

sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style ='smoker', s=100, ax=ax, palette=['#E24A33', '#348ABD'])
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)

## Se aplico el parametro framealpha para mostras la leyenda transparente y no tape los puntos
ax.legend(loc='upper right', fontsize='large', shadow=True, title='Detalles', framealpha=0)

# Añadir cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

# Añadir anotación
# Se obtiene los puntos maximos de cada columna del grafico, para ubicar el texto y flecha

x_max = df["total_bill"].max()
y_max = df["tip"].max()

# Punto de anotación al tope a la derecha
ax.annotate('Mayor tip en la cena', xy=(x_max, y_max), xytext=(x_max-10 , y_max+5),
            arrowprops=dict(facecolor='black', shrink=0.07))

# Añadir flecha
ax.arrow(x_max-21, y_max-5, 10, 3, head_width=0.8, head_length=0.5, fc='green', ec='green')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
