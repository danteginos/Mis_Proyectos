####  6.- Análisis de Datos

* Realiza agrupaciones de datos utilizando groupby para obtener insights específicos (por ejemplo, ventas por producto, ventas por región, etc.).

```python
 # Se muestra las sucursales

 df["sucursal"].value_counts().sort_values()
```

* Se deduce que hay 4 sucursales
* Se deduce sucursal con mayor cantidad de resgistros es 'Avenida Colón 608'

```python
# Ventas por turno y sucursal

df.groupby(["sucursal", "turno"])["total_venta"].sum().sort_values(ascending=False)
```

* Se deduce en la sucursal 'Avenida Colón 608' en turno 'Noche' tiene la mayores ventas

```python
 # Calcular total de clientes que son miembros con descuento miembro y los que no son miembros.

df.groupby(["cliente_miembro"])["descuento_miembro"].size()
```

* Se deduce que la cantidad de clientes miembros con descuento de miembro son 2450
* Se deduce que la cantidad de clientes sin membresia ni descuento de miembro son 2550

```python
# Se calcula ventas promedio por sucursal.

vta_promedio = df.groupby(df["sucursal"])["total_venta"].mean().round(2).sort_values(ascending=False)
print(vta_promedio)
```

Se deduce que la sucursal con el mayor promedio de ventas es la sucursal 'Avenida Colón 608' con 5300.97

```python
# Se calcula promedio de unidades vendidas de productos, por sucursal y categoria de producto.

df.groupby(["sucursal", "categoría_producto", "nombre_producto"])["cantidad"].mean().round(2).sort_values(ascending=False)
```

Se deduce que el promedio mas alto de unidades vendidas fue el productos 'Cold Brew' en la sucursal de 'Calle San Lorenzo 47' con valor de 3.67

```python
# Se Aplica el metodo apply para redondear a entero los valores de la columna 'demora_preparación_min'

df["demora_preparación_min"] = df["demora_preparación_min"].apply(lambda x: round(x))
```
